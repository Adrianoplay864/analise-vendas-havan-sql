```python
import pandas as pd
import matplotlib.pyplot as plt

# Create DataFrame
data = [
    ('2025-11', 203276.75),
    ('2025-03', 508421.08),
    ('2024-04', 513916.80),
    ('2024-10', 532838.72),
    ('2025-09', 536847.18),
    ('2025-08', 537348.15),
    ('2024-11', 551506.66),
    ('2025-06', 560548.58),
    ('2025-04', 561673.57),
    ('2024-08', 562764.97),
    ('2024-07', 575669.04),
    ('2024-03', 597807.20),
    ('2024-09', 599248.37),
    ('2025-07', 606107.33),
    ('2025-01', 607287.00),
    ('2025-02', 620568.26),
    ('2024-01', 621651.13),
    ('2025-10', 623854.42),
    ('2024-02', 627758.34),
    ('2025-05', 652108.38),
    ('2024-05', 677513.78),
    ('2024-12', 726271.33),
    ('2024-06', 761240.69)
]

df = pd.DataFrame(data, columns=["mes", "faturamento_total"])
df['mes'] = pd.to_datetime(df['mes'], format='%Y-%m')
df = df.sort_values('mes')

# Plot
plt.figure(figsize=(10, 5))
plt.plot(df['mes'], df['faturamento_total'])
plt.xlabel("Mês")
plt.ylabel("Faturamento Total")
plt.title("Faturamento por Mês")
plt.xticks(rotation=45)
plt.tight_layout()

# Save file in the current folder
plt.savefig("faturamento_por_mes.png")

"Arquivo salvo com sucesso!"
```
